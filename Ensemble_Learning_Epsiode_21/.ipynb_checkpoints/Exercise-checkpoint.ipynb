```python
import pandas as pd

df = pd.read_csv("heart (1).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import numpy as np
from scipy import stats
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
z_scores = abs(stats.zscore(df[numeric_cols]))

tol = (z_scores < 3).all(axis = 1)
df1 = df[tol].reset_index(drop = True)
df1
```

```python
cat_columns = ['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']
new_df = pd.get_dummies(df1, columns = cat_columns , prefix = cat_columns, drop_first = True)
new_df
```

```python
X = new_df.drop(['HeartDisease'], axis = 1)
y = new_df.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(), X, y , cv = 5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator = SVC(),
    n_estimators = 100,
    max_samples = 0.8,
    random_state = 0
)

scores = cross_val_score(bag_model, X,y, cv = 5)
scores.mean()
```

```python
# As you can see above, using baggin in SVM won't makea. difference
# it only makes a difference when we have high variance and instable model such as decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier

scores = cross_val_score(DecisionTreeClassifier(), X,y, cv = 5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator = DecisionTreeClassifier(),
    n_estimators = 100,
    max_samples = 0.8,
    random_state = 0
)

scores = cross_val_score(bag_model, X,y, cv = 5)
scores.mean()
```

```python
#Score improved from .72 to .79
```

```python
# Training a model using Random Forest which itself uses bagging
```

```python
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(),X,y, cv = 5)
scores.mean()
```
