```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("test_scores.csv")
df
```

```python
def gradient_descent(x,y):
    
    m_curr = b_curr = 0
    iterations = 100000
    n = len(x)
    learning_rate = 0.0002
    prev_cost = float("inf")
    tol = 1e-8
    
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr
        cost = (1/n) *sum([val**2 for val in (y-y_predicted)])
        
        if math.isclose(prev_cost , cost , rel_tol = tol):
            print(f"Converged (Δcost < tol) at iteration {i}")
            break
            
        prev_cost = cost
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        if i % 100 == 0: 
            print(f"iter {i:6d}  m {m_curr:.4f}  b {b_curr:.4f}  cost {cost:.4f}")
        pass

x = df['math'].to_numpy(dtype = "float")
y = df['cs'].to_numpy(dtype = "float")

gradient_descent(x,y)
# first reduce the iterations and change the learning rate and when you find a good learning rate
#then increase the iterations
# means “stop when the relative change in cost is less than 10⁻⁸ (i.e. 0.000001%) between iterations.”
# So an MSE of 31.6 means that, on average, each point’s squared error is 31.6.
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics    import mean_squared_error

r = LinearRegression().fit(df[["math"]], df["cs"])
preds = r.predict(df[["math"]])
mean_squared_error(df["cs"], preds)   # ≈ 31.6
```
